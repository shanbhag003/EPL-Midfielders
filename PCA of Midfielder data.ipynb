## 2) Sorting the Dataset

```python
#Independent Features
X = df.drop(["Pos",'Player', 'Squad',], axis = 1)
print(X.shape)

#Target
y = df["Pos"]
print(y.shape)
```

```python
feat_names = X.columns
print(feat_names)

target_names = ['DF,MF', 'MF', 'FW,MF']
print(target_names)
```

Now, we have sorted the data into Independent and target feature. We can sxale the data using MinMaxScaler (Min value = 0 and Max value = 1 before applying PCA.

## 3) Applying Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
```

```python
X_reduced = pca.fit_transform(X_scaled)
```

```python
print("Shape of Original dataset {}".format(X.shape))
print("Shape of reduced dataset due to PCA {}".format(X_reduced.shape))
```

#### We can see, the dimensions have reduced from 16 to 2.

## 4) Visualization

```python
import plotly.express as px
fig = px.scatter(X_reduced, x=0, y=1, color=df['Pos'], title = "PCA of EPL Midfielders dataset")
fig.show()
```
